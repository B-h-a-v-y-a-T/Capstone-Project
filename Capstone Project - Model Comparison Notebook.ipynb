```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
from datetime import datetime
import pathway as pw
import bokeh.plotting
import panel as pn
```

```python
#Upload the 2 csv(mentioned) files before runnning this cell

df1 = pd.read_csv("Model1_Output.csv")
df2 = pd.read_csv("Model2_Output.csv")
```

```python
#I renamed the columns to differentiate between the models

df1.rename(columns={"Price": "Price_Model_1"},inplace=True)
df2.rename(columns={"Price": "Price_Model_2"}, inplace=True)
```

```python
df= pd.merge(df1, df2, on=["Timestamp", "SystemCodeNumber"])

df['Timestamp']=pd.to_datetime(df['Timestamp'])
```

```python
pn.extension()

def comparison_plot(df):

  spots=df["SystemCodeNumber"].unique()
  subplots=[]

  for spot in spots:
    subset=df[df["SystemCodeNumber"]==spot]

    fig = bokeh.plotting.figure(
        x_axis_type="datetime",
        width=800,
        height=800,
        title=f"Model 1 vs Model 2 Pricing Comparison for Parking Spot: {spot}"
    )

    fig.line(subset["Timestamp"], subset["Price_Model_1"], line_width=2, color="navy", legend_label ="Model 1" )
    fig.scatter(subset["Timestamp"], subset["Price_Model_1"], size=3, color="red")

    fig.line(subset["Timestamp"], subset["Price_Model_2"], line_width=2, color="yellow",legend_label="Model 2")
    fig.scatter(subset["Timestamp"], subset["Price_Model_2"], size=3, color="green")

    fig.legend.location = "top_left"

    subplots.append(fig)

  return pn.Column(*subplots)

pn.panel(comparison_plot(df)).servable()
```
